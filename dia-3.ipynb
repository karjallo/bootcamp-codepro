```python
palabra_secreta = list(input('ingrese la palabra secreta'))
cantidad_intentos = int(input('ingrese la cantidad de intentos'))

cantidad_letras = len(palabra_secreta)
intentos_restantes = cantidad_intentos

# solo vuelvo a definir aca, porque jupyter traeria el valor de otro bloque de codigo
# o de un intento anterior
cantidad_aciertos = 0
resultado = []

# crear funcion de verificacion de palabras
def intento_valido():
    while True:
        palabra = input(f'ingrese su palabra de {cantidad_letras} caracteres, Intentos restantes:{intentos_restantes}')
        if len(palabra) != cantidad_letras:
            print(f'la palabra propuesta, no cuenta con {cantidad_letras} caracteres, vuelva a ingresar otra palabra')
        else:
            return list(palabra)


# iteramos haciendo el mismo procedimiento
for i in range(cantidad_intentos): 
    # llamamos a la funcion para validar el intento
    palabra_ingresada = intento_valido()
    # reiniciamos valores de las variables para cada ciclo
    cantidad_aciertos = 0
    resultado = []
    
    # iteramos sobre cada letra para comprobar acierto total, parcial
    # o no acierto y hacemos append al resultado
    for j in range(cantidad_letras):
        if palabra_ingresada[j] == palabra_secreta[j]:
            resultado.append('[' + palabra_ingresada[j] + ']' )
            cantidad_aciertos = cantidad_aciertos + 1
        elif palabra_ingresada[j] in palabra_secreta:
            resultado.append('('+ palabra_ingresada[j] +')')
        else:
            resultado.append(palabra_ingresada[j])

    # bajamos contador de intentos restantes
    intentos_restantes = intentos_restantes - 1
    # hacemos print del resultado
    print(resultado)
    
    # generamos un break para salir en caso del acierto o sin intentos
    if cantidad_aciertos == cantidad_letras:
        print('ganaste capo, master, pantera, emperador')
        break
    elif intentos_restantes == 0:
        print('perdiste, perdiste, no hay nadie peor que vos')
        break
```
