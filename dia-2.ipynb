Challenge  - 20m

Escribe un programa que pregunte al usuario su edad para ingresar al festival Asunciónico y determine qué tipo de acceso puede tener.

Si el usuario tiene menos de 18 imprimir: "Acceso denegado — Solo zona Kids "

Si el usuario tiene mas de 18 pero menos de 21 imprimir: "Acceso permitido — No puede comprar bebidas alcoholicas "

Si el usuario tiene mas de 21 imprimir: "Acceso total — Disfruta el festival "

```python
edad = int(input('cual es su edad?:'))

if edad < 18:
    print('Acceso denegado - Solo zona kids')
elif edad < 21:
    print('Acceso permitido - No puede comprar bebidas alcoholicas')
else:
    print("Acceso total - Disfruta el festival")
```

Challenge  - 30m,
Escribir un programa en el que se pregunte al usuario por una temperatura (en grados Celsius), imprima la temperatura escrita, y luego clasificar la temperatura ingresada en los siguientes niveles:

Entre -10 y 13 grados, que imprima "Ro'y koko".,
Entre 14 y 23 grados, que imprima "Refresca'i".,
Entre 24 y 28 grados, que imprima "Oikoite".,
Entre 29 y 45 grados, que imprima "Hakurooooo".,
Para todo lo demás, imprimir "Ñamano mba'e!".

```python
temperatura = int(input('cual es la temperatura en grados Celsius?'))
print(f"la temperatura es: {temperatura}")

if temperatura >= -10 and temperatura <= 13:
    print("Ro'y koko")
elif temperatura >= 14 and temperatura <= 23:
    print("Refresca'i")
elif temperatura >= 24 and temperatura <= 28:
    print("Oikoite")
elif temperatura >= 29 and temperatura <= 45:
    print("Hakurooooo")
else:
    print("Ñamano mba'e!")
```

Challenge 🤺 - 20m
Crear una lista de 5 elementos. Luego, seguir los siguientes pasos.

Agregar un elemento al final de la lista.

Modificar el segundo elemento.

Eliminar el tercer elemento.

```python
lista_ingredientes = ['lechuga', 'tomate', 'pan', 'carne', 'cebolla']
print(f'''La lista original es:
{lista_ingredientes}
''')

lista_ingredientes.append('carbon')
print(f'''Agregamos al final de la lista:
{lista_ingredientes}
''')

lista_ingredientes[1] = 'bacon'
print(f'''Modificamos el segundo elemento:
{lista_ingredientes}
''')

lista_ingredientes.pop(2)
print(f'''Eliminamos el tercer elemento:
{lista_ingredientes}
''')
```

Challenge 🤺 - 20m
Recorrer los elementos de una lista de marcas e imprimir cada marca solamente si tiene 6 letras o más.

```python
lista_marcas = ['adidas', 'nike', 'volkswagen', 'toyota', 'philips', 'panasonic']

for marca in lista_marcas:
    if len(marca) >= 6:
        print(marca)

```

Challenge  - 15m,
Solicitar al usuario que ingrese una frase.

Imprimir un mensaje cada vez que se encuentre una letra "o".

```python
frase_usuario = input('ingrese una frase:')

for caracter in frase_usuario:
    # print(caracter)
    if caracter == 'o' or caracter == 'O':
        print(f' letra {caracter} detectada')
```
