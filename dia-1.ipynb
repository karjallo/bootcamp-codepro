Challenge 🤺 – 15m

🎯 Objetivo: Practicar división y módulo.

Imagina que estás repartiendo caramelos: solicita al usuario ingresar el total de caramelos y la cantidad de niños, luego calcula cuántos caramelos recibe cada niño usando división entera y cuántos sobran usando módulo, e imprime ambos resultados.

```python
total_caramelos = int(input("cual es la cantidad de caramelos?"))
total_ninos = int(input("cuantos ninos hay?"))

caramelos_por_nino =  total_caramelos // total_ninos
caramelos_restantes =  total_caramelos % total_ninos

print(f"cada nino recibe {caramelos_por_nino} caramelos y sobran {caramelos_restantes} caramelos")
```

Challenge  - 10m

 Objetivo:
Practicar variables globales y locales.

Escribir una función calcular_total que reciba el precio de un producto y calcule el total sumando un impuesto global impuesto = 10.
Usa una variable local total para almacenar el resultado e imprime el total.
Llama a la función con distintos precios para probar.

```python
impuesto = 0.1

def calcular_total(precio):
    costo_impuesto = precio*impuesto
    precio_final = precio + costo_impuesto
    print(f'el precio total del producto es: {precio_final}')

precio_producto = int(input("cual es el valor del producto?:"))
calcular_total(precio_producto)
```
