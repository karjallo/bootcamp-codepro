Challenge 1 🤺
Escribir un programa que cree un diccionario vacío y lo vaya llenado con información sobre una persona (por ejemplo nombre, edad, sexo, teléfono, correo electrónico, etc.) que se le pida al usuario. 

Cada vez que se añada un nuevo dato debe imprimirse el contenido del diccionario.

```python

mi_diccionario = {}

datos_a_solicitar = ['nombre', 'apellido', 'telefono', 'direccion', 'barrio']


# datos_a_solicitar[i] ira tomando el valor de la lista datos a solicitar
for i in range(len(datos_a_solicitar)):
    mi_diccionario[datos_a_solicitar[i]] = input(f'ingrese su {datos_a_solicitar[i]}')
    print(mi_diccionario)

```

Challenge 2 🤺
Crear una tarjeta personal. El usuario (mediante input) puede elegir:

Crear datos nuevos
Modificar datos
Eliminar datos

```python
tarjeta_personal = {}

def crear_dato():
    llave = input("que dato desea agregar:?")
    tarjeta_personal[llave] = input("cual es el valor?:")
    print(f'tarjeta personal actualizada: \n {tarjeta_personal}')

def modificar_dato():
    llave = input(f"que valor desea modificar? {tarjeta_personal.keys()}")
    if llave in tarjeta_personal.keys():
        tarjeta_personal[llave] = input("cual es el valor?:") 
        print(f'tarjeta personal actualizada: \n {tarjeta_personal}')
    else:
        print('dato no encontrado, ingrese nuevamente')
    
def eliminar_dato():
    llave = input(f'que dato desea eliminar? {tarjeta_personal.keys()}')
    if llave in tarjeta_personal.keys():
        del tarjeta_personal[llave]
        print(f'tarjeta personal actualizada: \n {tarjeta_personal}')
    else:
        print('dato no encontrado, ingrese nuevamente')


while True:
    accion_a_realizar = input("Que desea hacer?: 1 - crear, 2 - modificar o  3- eliminar un dato? \n presione cualquier otro valor para salir")
    if accion_a_realizar == '1':
        crear_dato()
    elif accion_a_realizar == '2':
        modificar_dato()
    elif accion_a_realizar == '3':
        eliminar_dato()
    else:
        print(f'su tarjeta personal es')
        print(tarjeta_personal)
        print('hasta luego')
        break



```

Challenge 3 🤺
Crear una agenda de contactos. El usuario puede elegir:

Mostrar los contactos
Crear contacto
Modificar datos del contacto

```python
agenda_contactos = {
    'angel' : {'numero':'0985111222', 'mail':'angel@gmail.com'},
    'samuel' : {'numero': '0981999000', 'mail':'samuel@gmail.com'},
}

def crear():
    nombre = input("ingrese el nombre de la persona")
    numero = input("cual es el numero de telefono?:")
    mail = input("cual es su mail?:")
    agenda_contactos[nombre] = {'numero':numero, 'mail':mail}
    print(f'agenda actualizada: \n {agenda_contactos}')

def modificar():
    nombre = input(f"cual contacto desea modificar? {agenda_contactos.keys()}")
    if nombre in agenda_contactos.keys():
        numero = input("cual es el numero de telefono?:")
        mail = input("cual es su mail?:")
        agenda_contactos[nombre] = {'numero':numero, 'mail':mail}
        print(f'agenda actualizada: \n {agenda_contactos}')
    else:
        print('nombre no encontrado, ingrese nuevamente')
    
def mostrar():
    print(agenda_contactos)

while True:
    accion_a_realizar = input("Que desea hacer?: 1 - crear, 2 - modificar o  3- mostrar contactos? \n presione cualquier otro valor para salir")
    if accion_a_realizar == '1':
        crear()
    elif accion_a_realizar == '2':
        modificar()
    elif accion_a_realizar == '3':
        mostrar()
    else:
        print('hasta luego')
        break
```
