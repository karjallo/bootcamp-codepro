### Ejercicio 1 - 7 min (F)
Vamos a usar un bucle `while` para mostrar números en pantalla.

Objetivo:
Mostrar los números del **1 al 5**, uno por uno.

💡 Pista:
- Necesitas una variable contador.
- El bucle debe repetirse mientras el número sea menor o igual a 5. 

```python
i = 1
while i <= 5:
    print(i)
    i = i + 1
```

Objetivo:
Pedir al usuario un número.
El programa debe seguir pidiendo mientras el número sea distinto de 10.

Cuando el usuario escriba 10, mostrar:
"¡Correcto!"

```python
while True:
    numero = input('ingrese el numero 10')
    if (numero == '10'):
        print('Correcto')
        break
    else:
        print('numero incorrecto, ingrese nuevamente')
```

Challenge 🤺 - 15 min
Definir una variable que contenga un numero y preguntar al usuario por el numero hasta que acierte.
El programa debe tener una cantidad limitada de intentos.
Si falla imprimir si el numero a adivinar es mayor o menor que el numero ingresado y debe indicar la cantidad de intentos que le quedan.
Si acierta imprimir un mensaje.

```python
numero = 91
intentos = 5

while intentos > 0:
    numero_ingresado = int(input(f'ingrese un numero, intentos restantes {intentos}'))
    if numero_ingresado == numero:
        print(f'ganaste, el numero era {numero_ingresado}')
        break
    elif numero_ingresado < numero:
        print('el numero secreto es mayor')
        intentos = intentos - 1
    else:
        print('el numero secreto es menor')
        intentos = intentos -1

if intentos == 0:
    print('te quedaste sin intentos, perdiste')


```

Challenge  - (F) 12 min,
Objetivo:
Crear una función llamada compra de producto.,
Debe recibir 3 parámetros:
1) nombre
2) producto
3) cantidad,
La función debe mostrar un mensaje usando esos datos.

```python
def compra_de_producto(nombre, producto, cantidad):
    print(f'{nombre} ha comprado {cantidad} unidades de {producto}')

compra_de_producto('Oscar', 'chocolate', 15)
```

## ⚔️ Challenge  — 5 min
La lista de animales

Un zoológico quiere mostrar todos sus animales.

🎯 Tu misión

Crear una lista con 5 animales

Usar for para imprimir cada animal

```python
lista_animales = ['jaguarete', 'jirafa', 'elefante', 'rinoceronte', 'avestruz']

for i in range(len(lista_animales)):
    print(lista_animales[i])

```
