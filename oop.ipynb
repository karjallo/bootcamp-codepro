Challenge 🤺
Crear dos métodos para nuestro animal y ejecutar, debe ser algo específico de ese animal.

Ejemplo: "Cuando hablo, digo X".
Ejemplo: "Cuando me muevo, hago X".

```python
class Animales:
    especie = 'vacas'

    def __init__(self, nombre, posicion=0):
        self.nombre = nombre
        self.posicion = posicion

    def correr(self, metros):
        self.metros = metros
        self.posicion = self.posicion + self.metros
        print(f'la vaca {self.nombre} corrio {self.metros}')
    
    def escapa(self):
        if self.posicion > 50:
            print(f'la vaca {self.nombre} logro escapar')
        else:
            print(f'la vaca {self.nombre} no pudo escapar :c')

vaca1 = Animales('lola')
print(vaca1.nombre)

vaca1.correr(40)
vaca1.escapa()
vaca1.correr(20)
vaca1.escapa()
```

Challenge 🤺15 min
Definir una clase madre que herede a sus clases hijas dos atributos de instancia. Mostrar ambos atributos en pantalla desde los objetos de las clases hijas.


```python
class Ejercito:
    tipo = 'milicia'
    agua = False
    tierra = False
    aire = False
    
    def __init__(self, nombre, antiguedad):
        self.nombre = nombre
        self.antiguedad = antiguedad
    def presentarse(self):
        print(f'mi nombre es {self.nombre}, pertenezco a la {self.tipo}, tengo {self.antiguedad} anhos de antiguedad')

class Naval(Ejercito):
    agua = True
    
class Infanteria(Ejercito):
    tierra = True

class Aereo(Ejercito):
    aire = True

naval1 = Naval('jorge',5)
naval1.presentarse()
print(naval1.agua)

infanteria1 = Infanteria('juan', 10)
infanteria1.presentarse()
print(infanteria1.tierra)

aereo1 = Aereo('jose', 4)
aereo1.presentarse()
print(aereo1.aire)
```

Challenge 🤺
Crear dos personajes de una historia que se representen a traves de clases. Ejemplo: Un hereo y un villano.
Cada personaje debe tener poderes o acciones diferentes. Y deben presentarse en una frase.

Ejemplo: Yo me llamo pepito y tengo super fuerza
Ejemplo: Yo me llamo juanito y puedo volar

```python
class Humano():
    def __init__(self, nombre, poder):
        self.nombre = nombre
        self.poder = poder
    def saludo(self):
        print(f'Hola soy {self.nombre} y mi poder es {self.poder}')
        

class Heroe(Humano):
    caracteristica = 'altruista'
    def defender(self):
        print(f'con el poder de {self.poder} ayudo a traer justicia al pueblo')


class Villano(Humano):
    caracteristica = 'maligno'
    def atacar(self, cantidad):
        print(f'con mi {self.poder}, causo caos en la ciudad, matando asi a {cantidad} de personas')


heroe1 = Heroe('juan','volar')
heroe1.saludo()
print(heroe1.caracteristica)
heroe1.defender()

villano1 = Villano('jorge','superfuerza')
villano1.saludo()
print(villano1.caracteristica)
villano1.atacar('10')
```
